```python
#split the data into train and test
x_train, x_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(x_train)
```

```python
 # encode the dataframe
 from sklearn.preprocessing import LabelEncoder
 le = LabelEncoder()
 iris['Species'] = le.fit_transform(iris['Species'])
```

```python
# train the model
model = LogisticRegression()
model.fit(x_train, Y_train)
```

```python
 # predict the model
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
 # save model using joblib
import joblib
 joblib.dump(model, 'iris_model.joblib')
```

```python
 # predict flower using saved model
 model = joblib.load('iris_model.joblib')
 predict = model.predict([[5.1,3.5,1.4,0.2]])
 predict[0]
```

```python
 # accuracy of model
 accuracy = accuracy_score(Y_test, y_pred)
 print('Accuracy:', accuracy)
```

```python
 # confusion matrix
 from sklearn.metrics import confusion_matrix
 cm = confusion_matrix(Y_test, y_pred)
 print(cm)
```

```python
 # plot confusion matrix graph
 import seaborn as sns
 sns.heatmap(cm, annot=True)
 plt.show()
```

```python
 # plot actual and predicted
 plt.scatter(Y_test, y_pred)
 plt.show()
```

```python
le=LabelEncoder()
yy=le.fit_transform(Y_test)
xx=le.fit_transform(y_pred)
```

```python
 # calculate mean_squared_error, r2_score
mse = mean_squared_error(yy,xx)
r2 = r2_score(yy,xx)
print('Mean Squared Error:', mse)
print('R2 Score:', r2)
```
